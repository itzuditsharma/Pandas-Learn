Pivoting “Long” to “Wide” Format

```python
stock = pd.DataFrame({
    "fruit" : ["apple", "plum", "grape"] * 2,
    "color" : ["purple", "yellow"] * 3,
    "piece" : [3,4,5,6,1,2]
})

stock
```

```python
stock.pivot(index = "fruit", columns="color", values = "piece")
```

```python
stock
```

```python
stock["values"] = np.random.randn(len(stock))
stock
```

```python
p = stock.pivot(index="fruit", columns="color")
p
```

```python
p["values"]
```

Pivoting “Wide” to “Long” Format

```python
data=pd.DataFrame(
    {"lesson":["math","stat","bio"],
     "Sam":[50,60,70],
     "Kim":[80,70,90],
     "Tom":[60,70,85]})
data
```

```python
group = pd.melt(data, ["lesson"])
group
```

```python
n = group.pivot(index = "lesson", columns="variable", values = "value")
n
```
